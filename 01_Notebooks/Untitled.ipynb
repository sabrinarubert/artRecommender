```python
sqlEngine       = create_engine('mysql+pymysql://root:sabrina_art@127.0.0.1:3301/sabrina_art')
dbConnection    = sqlEngine.connect()
```

```python
allArtworks = pd.read_sql("select id as artwork_id from artworks LIMIT 100", dbConnection)
```

```python
for index, row in allArtworks.iterrows():
    
    artwork_id = row["artwork_id"]    
    genes_table = pd.read_sql("SELECT * FROM artworks_to_genes WHERE artwork_id LIKE '"+ artwork_id +"'", dbConnection)
    
    for index2, gene in genes_table.iterrows():
        
        gene_id = gene["gene_id"]
        allArtworks.loc[index, gene_id] = 1
        
        
allArtworks = allArtworks.fillna(0)
allArtworks.head(100)
```
